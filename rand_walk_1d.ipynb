# random walk 1-d

## mean distance

```python
import random as rnd
import matplotlib.pyplot as plt

def mean_displacement_series(Nrep, Nstep):
    # Menghasilkan rata-rata posisi/ displacement pada tiap step (mengikuti pola kode awal)
    sx = [0.0] * Nstep
    for _ in range(Nrep):
        x = [0] * Nstep
        for i in range(Nstep):
            d = 2 * rnd.randint(0, 1) - 1  # +1 atau -1
            x[i] = d
        for i in range(Nstep):
            sx[i] += x[i]
    for i in range(Nstep):
        sx[i] /= Nrep
    return sx

# contoh awal (sesuai template yang kamu kasih)
Nrep = 100
Nstep = 21
sx = [0] * Nstep
for j in range(Nrep):
    x = [0] * Nstep
    for i in range(0, Nstep):
        d = 2*rnd.randint(0, 1) - 1
        x[i] = d
    for i in range(0, Nstep):
        sx[i] = sx[i] + x[i]
for i in range(0, Nstep):
    sx[i] = sx[i] / Nrep

plt.figure()
plt.plot(sx)
plt.xlabel('steps')
plt.ylabel('average distance (mean displacement)')
plt.grid()
plt.show()
```

## mean squared distance

```python
import random as rnd

def mean_squared_distance_series(Nrep, Nstep):
    sx2 = [0.0] * Nstep
    for _ in range(Nrep):
        pos = 0
        for i in range(Nstep):
            d = 2 * rnd.randint(0, 1) - 1
            pos += d
            sx2[i] += pos * pos
    for i in range(Nstep):
        sx2[i] /= Nrep
    return sx2

Nrep = 10
Nstep = 121
sx2 = [0] * Nstep

# generate steps for random walk
for j in range(Nrep):
    pos = 0
    for i in range(0, Nstep):
        d = 2*rnd.randint(0, 1) - 1
        pos = pos + d
        sx2[i] = sx2[i] + pos * pos

# mean squared distance for Nrep
for i in range(0, Nstep):
    sx2[i] = sx2[i] / Nrep

import matplotlib.pyplot as plt
plt.plot(sx2)
plt.xlabel('steps')
plt.ylabel('mean squared distance')
plt.grid()
plt.show()
```

## Random Walk 1D

### Mean Distance (Rata-Rata Posisi)
Pada *random walk* satu dimensi, partikel bergerak acak ke kanan (+1) atau kiri (–1) dengan peluang sama.  
Secara teori, rata-rata posisinya selalu nol:
$$
\langle x_n \rangle = 0
$$
karena sistem simetris — jumlah langkah ke kanan dan kiri seimbang.  
Dalam simulasi dengan `Nrep` kecil, hasil bisa menyimpang dari nol karena efek acak.  
Semakin besar `Nrep`, hasil rata-rata makin halus dan mendekati nilai teorinya sesuai **Hukum Bilangan Besar**.

---

### Mean Squared Distance (Jarak Kuadrat Rata-Rata)
Besarnya jarak kuadrat rata-rata dari titik asal adalah:
$$
\langle x_n^2 \rangle = n
$$
yang berarti pertumbuhannya **linear terhadap jumlah langkah**.  
Ini menunjukkan pola **difusi normal**, di mana partikel menyebar lebih jauh seiring waktu.  
Jarak rata-rata dari titik asal berbanding dengan akar jumlah langkah:
$$
\langle |x_n| \rangle \propto \sqrt{n}
$$
Semakin banyak langkah yang diambil, partikel makin jauh dari posisi awal, tetapi dengan pertumbuhan yang melambat (akar kuadrat, bukan linear).
