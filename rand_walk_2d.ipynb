# random walk 2-d

## mean distance

```python
import random as rnd
import matplotlib.pyplot as plt

Nrep = 50
Nstep = 101

mean_r = [0.0] * Nstep

# simulate random walks
for j in range(Nrep):
    x = 0.0
    y = 0.0
    for i in range(Nstep):
        # random step: one of four directions (N, S, E, W)
        direction = rnd.randint(0, 3)
        if direction == 0:
            x += 1
        elif direction == 1:
            x -= 1
        elif direction == 2:
            y += 1
        else:
            y -= 1

        r2 = x*x + y*y
        mean_r[i] += r2**0.5      # accumulate distance

# average over repetitions
for i in range(Nstep):
    mean_r[i] /= Nrep

# Plot results
plt.figure(figsize=(8,5))
plt.plot(mean_r)
plt.xlabel('steps')
plt.ylabel('mean distance')
plt.grid(True)
plt.show()
```

## Random Walk 2D

### Mean Distance (Rata-Rata Jarak)
Pada *random walk* dua dimensi, partikel bisa bergerak ke kanan, kiri, atas, atau bawah dengan peluang yang sama besar.  
Posisi partikel setelah `n` langkah adalah \((x_n, y_n)\), dan jaraknya dari titik awal:
$$
r_n = \sqrt{x_n^2 + y_n^2}
$$
Secara teori, jarak rata-rata partikel terhadap titik asal meningkat sebanding dengan akar jumlah langkah:
$$
\langle r_n \rangle \propto \sqrt{n}
$$
Artinya, semakin banyak langkah yang diambil, jarak rata-rata partikel makin besar, tapi pertumbuhannya melambat (bukan linear).  
Semakin besar `Nrep`, hasil simulasi akan semakin halus dan mendekati bentuk teorinya karena efek acak saling menetralkan.

---

### Mean Squared Distance (Jarak Kuadrat Rata-Rata)
Untuk 2D, jarak kuadrat rata-rata dari titik asal diberikan oleh:
$$
\langle r_n^2 \rangle = \langle x_n^2 + y_n^2 \rangle = 2D\,n
$$
dengan \(D\) adalah **koefisien difusi**.  
Dalam simulasi sederhana dengan langkah ±1, nilai \(D \approx \tfrac{1}{2}\), sehingga secara numerik:
$$
\langle r_n^2 \rangle \approx n
$$
Jadi, pertumbuhan *mean squared distance* (MSD) tetap **linear terhadap jumlah langkah**, sama seperti di 1D, hanya saja nilainya lebih besar karena ada dua arah gerak.

---

### Kesimpulan
- Rata-rata jarak meningkat seperti $$\sqrt{n}$$  
- Rata-rata jarak kuadrat meningkat linear seperti $$n$$  
- Nilai `Nrep` yang lebih besar membuat hasil simulasi lebih halus dan mendekati nilai teoritis karena variasi acak berkurang.
