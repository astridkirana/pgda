# random walk n-d

## assignment 5
1. Explain the influence of `Nrep` to value of `mean distance` in 1-d random walk.
2. Explain the influence of `Nrep` to value of `mean distance` in 2-d random walk.
3. Explain the influence of `Nrep` to value of `mean squared distance` in 1-d random walk.
4. Explain the influence of `Nrep` to value of `mean squared distance` in 2-d random walk.
5. Summerize and generalize the influence of `Nrep` to value of `mean distance` and `mean squared distance` in n-d random walk.

```python
import random as rnd
import matplotlib.pyplot as plt

def mean_distance_series_nd(Nrep, Nstep, ndim):
    s = [0.0] * Nstep
    for _ in range(Nrep):
        pos = [0] * ndim
        for i in range(Nstep):
            axis = rnd.randint(0, ndim - 1)
            step = 1 if rnd.randint(0, 1) == 1 else -1
            pos[axis] += step
            # hitung ||pos||
            r2 = 0
            for v in pos:
                r2 += v * v
            s[i] += r2 ** 0.5
    for i in range(Nstep):
        s[i] /= Nrep
    return s

def mean_squared_distance_series_nd(Nrep, Nstep, ndim):
    s2 = [0.0] * Nstep
    for _ in range(Nrep):
        pos = [0] * ndim
        for i in range(Nstep):
            axis = rnd.randint(0, ndim - 1)
            step = 1 if rnd.randint(0, 1) == 1 else -1
            pos[axis] += step
            r2 = 0
            for v in pos:
                r2 += v * v
            s2[i] += r2
    for i in range(Nstep):
        s2[i] /= Nrep
    return s2

Nstep = 101
Nreps = [50, 100, 200, 400, 800]

```


# 1D: mean distance

```python
md_1d = {}
for nr in Nreps:
    md_1d[nr] = mean_distance_series_nd(nr, Nstep, ndim=1)

plt.figure()
for nr in Nreps:
    plt.plot(md_1d[nr], label=f'Nrep={nr}')
plt.xlabel('steps')
plt.ylabel('mean distance (1D)')
plt.grid(True)
plt.legend()
plt.show()
```

# 2D: mean distance

```python
md_2d = {}
for nr in Nreps:
    md_2d[nr] = mean_distance_series_nd(nr, Nstep, ndim=2)

plt.figure()
for nr in Nreps:
    plt.plot(md_2d[nr], label=f'Nrep={nr}')
plt.xlabel('steps')
plt.ylabel('mean distance (2D)')
plt.grid(True)
plt.legend()
plt.show()
```

# 1D: mean squared distance

```python
msd_1d = {}
for nr in Nreps:
    msd_1d[nr] = mean_squared_distance_series_nd(nr, Nstep, ndim=1)

plt.figure()
for nr in Nreps:
    plt.plot(msd_1d[nr], label=f'Nrep={nr}')
plt.xlabel('steps')
plt.ylabel('mean squared distance (1D)')
plt.grid(True)
plt.legend()
plt.show()
```

# 2D: mean squared distance

```python
msd_2d = {}
for nr in Nreps:
    msd_2d[nr] = mean_squared_distance_series_nd(nr, Nstep, ndim=2)

plt.figure()
for nr in Nreps:
    plt.plot(msd_2d[nr], label=f'Nrep={nr}')
plt.xlabel('steps')
plt.ylabel('mean squared distance (2D)')
plt.grid(True)
plt.legend()
plt.show()
```

# Penjelasan

## 1. Pengaruh `Nrep` terhadap Mean Distance di 1D
Pada random walk satu dimensi (1D), rata-rata posisi partikel secara teoritis bernilai nol:
$$
\langle x_n \rangle = 0
$$
Karena peluang langkah ke kanan dan kiri sama besar, hasil simulasi akan berfluktuasi di sekitar nol.  
Semakin besar `Nrep`, rata-rata dari banyak simulasi akan semakin mendekati nol karena fluktuasi acak saling meniadakan (sesuai *Law of Large Numbers*).  
Dengan kata lain, `Nrep` yang besar menghasilkan kurva **lebih halus** dan mendekati nilai teoritis.

---

## 2. Pengaruh `Nrep` terhadap Mean Distance di 2D
Dalam random walk dua dimensi, jarak partikel dari titik asal didefinisikan sebagai:
$$
r_n = \sqrt{x_n^2 + y_n^2}
$$
Secara teoritis, rata-rata jarak partikel mengikuti hubungan:
$$
\langle r_n \rangle \propto \sqrt{n}
$$
Dengan `Nrep` kecil, hasil simulasi bisa “kasar” karena variasi antar percobaan masih besar.  
Jika `Nrep` diperbesar, hasil rata-rata menjadi lebih stabil dan mendekati bentuk akar kuadrat yang diharapkan dari teori difusi.

---

## 3. Pengaruh `Nrep` terhadap Mean Squared Distance di 1D
Nilai ekspektasi jarak kuadrat rata-rata (MSD) untuk random walk 1D adalah:
$$
\langle x_n^2 \rangle = n
$$
MSD meningkat **linear terhadap jumlah langkah**.  
Ketika `Nrep` kecil, hasilnya bisa berfluktuasi; tetapi dengan `Nrep` besar, hasil rata-rata semakin mendekati garis linear \(n\), sesuai prediksi teori.

---

## 4. Pengaruh `Nrep` terhadap Mean Squared Distance di 2D
Untuk kasus dua dimensi, jarak kuadrat rata-rata dituliskan sebagai:
$$
\langle r_n^2 \rangle = \langle x_n^2 + y_n^2 \rangle = 2D\,n
$$
dengan \(D\) adalah koefisien difusi.  
Dalam simulasi sederhana, \(D\) ≈ ½ sehingga hubungan empiris tetap linear terhadap \(n\).  
Ketika `Nrep` diperbesar, penyimpangan dari garis linear berkurang dan hasil rata-rata menjadi lebih halus.

---

## 5. Generalisasi ke Random Walk n-Dimensional
Untuk ruang berdimensi \(d\), kedua hubungan utama tetap berlaku:
$$
\langle \mathbf{r}_n \rangle = \mathbf{0}
$$
$$
\langle r_n^2 \rangle = d\,n
$$
$$
\langle r_n \rangle \propto \sqrt{n}
$$
Artinya, peningkatan jumlah dimensi hanya mempengaruhi faktor skala (konstanta proporsionalitas), bukan bentuk pertumbuhannya.  
Sementara itu, peningkatan jumlah replikasi `Nrep` tidak mengubah tren teoritis, tetapi **mengurangi fluktuasi acak** dan membuat hasil simulasi:
- Lebih konsisten antar percobaan,  
- Lebih halus, dan  
- Lebih dekat dengan nilai ekspektasi teoritis.

---

## Kesimpulan Singkat
- `Nrep` berperan penting dalam memperhalus hasil simulasi.  
- Semakin besar `Nrep`, semakin kecil fluktuasi dan semakin akurat hasilnya dibanding teori.  
- Secara umum:
  - Mean distance → tumbuh seperti $$\sqrt{n}$$  
  - Mean squared distance → tumbuh linear terhadap $$n$$  
  - Efek `Nrep` → menstabilkan hasil tanpa mengubah tren utama.
